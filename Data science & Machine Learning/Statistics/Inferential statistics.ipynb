```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
```

```python
df=pd.read_csv('weight-height.csv')
df
```

# Sampling distribution:

```python
df['Height'].mean() #population mean
```

```python
s=df['Height'].sample(n=50) # to extract samples by simple random sampling
s
```

```python
s.mean() #sample mean
```

```python
sample_mean=[]
for i in range (1,1000):
    s=df['Height'].sample(n=50)
    sample_mean.append(s.mean())
```

```python
sample_mean
```

```python
plt.hist(sample_mean,bins=20)
plt.show()
#normal dist
```
